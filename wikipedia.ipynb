```python
def retrieve_inner_and_outer_length(country_page):
    
    country_title = country_page.title.text
    country_pagelength = len(country_page.mediawiki.page.text)
    
    text_to_parse = country_page.mediawiki.page.text
    re_inner_links = re.compile(r'\[\[.*?\]\]')
    inner_links = re_inner_links.findall(text_to_parse)
    inner_links_len = len(inner_links)
    
    re_outer_links = re.compile(r'https?://[\w\./?&=%]*')
    outer_links = re_outer_links.findall(text_to_parse)
    outer_links_len = len(outer_links)
                                         
    return [country_title, inner_links_len, outer_links_len, country_pagelength]
```

```python
import csv

scientist_names = [] 

with open('gscholar/files/profData_withGender_physics_economics_philo.csv', "r") as f:
    reader = csv.reader(f)

    # get the link loaded in the csv
    for row in reader:
        scientist_names.append(row[8])

scientist_names.remove('name')
len(scientist_names)
```

```python
def get_wiki_length(name_scientist):
    data = getWikiPage(name_scientist)
    
    global failed_names
    
    if data == 'failure':
        failed_names.append(name_scientist)
        pass
    
    try:
        soup = bs4.BeautifulSoup(data, "lxml")
    
        if type(soup.mediawiki.page) == type(None):
            wiki_exists = False
            wiki_length = 0
    
        else:
            wiki_exists = True
            wiki_length = len(soup.mediawiki.page.text)
            
            global counter
        counter = counter + 1
    
        if counter % 100 == 0 or counter == 1:
            print(counter, end="", flush=True)
    
        return [name_scientist, wiki_exists, wiki_length]
    
    except:
        print(name_scientist)
        failed_names.append(name_scientist)
        pass
    
    
```
